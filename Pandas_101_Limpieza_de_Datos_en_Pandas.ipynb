# Limpieza de Datos en Pandas

Referencia: https://www.youtube.com/watch?v=bDhvCp3_lYw&t=172s

```python
import pandas as pd
```

```python
# Cargamos el archivo
df = pd.read_excel("/content/Customer Call List.xlsx")
df
```

```python
df.info()
```

Es un dataset de (21, 8) o sea: 21 filas con 8 columnas. Observemos que algunas columnas presentan nulos como `Last_Name`, `Phone_Number`, `Do_Not_Contact`, ya que las columnas no concuerdan con el total de filas.

```python
df = df.drop_duplicates() # Elimina duplicados
df
```

```python
# La funciÃ³n de Pandas hace exactamente eso: unificar datos duplicados

data = {
    "A": ["TeamA", "TeamB", "TeamB", "TeamC", "TeamA"],
    "B": [50, 40, 40, 30, 50],
    "C": [True, False, False, False, True]
}

df_example = pd.DataFrame(data)

display(df_example.drop_duplicates())
```

```python
df = df.drop(columns = "Not_Useful_Column") # Elimina columnas
df
```

```python
df.shape
```

```python
# Si queremos remover caracteres sea al inicio o al final de un string, podemos utilizar la funciÃ³n "strip"
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df
```

```python
import re # LibrerÃ­a para expresiones regulares

df["Phone_Number"] = df["Phone_Number"].astype(str).apply(lambda x: re.sub(r'\D', '', x))
df
```

```python
# Dividimos

df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',', n=2, expand=True)
df
```

```python
# Estandarizamos las columnas que tengan valores booleanos

df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')

df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')

df
```

```python
df = df.replace('N/a','')
df = df.replace('NaN','')
df=df.fillna('')
df
```

```python
df.info()
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)
df
```

```python
df.shape
```

```python
df.info()
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)
df

# Otra forma de hacerlo
#df = df.dropna(subset="Phone_Number"), inplace=True)
```
